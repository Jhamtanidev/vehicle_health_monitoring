```python
import pandas as pd
import os
import librosa
import librosa.display
import numpy as np
import IPython.display as ipd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from itertools import cycle

sns.set_theme(style="white", palette=None)
color_pal = plt.rcParams["axes.prop_cycle"].by_key()["color"]
color_cycle = cycle(plt.rcParams["axes.prop_cycle"].by_key()["color"])
```

```python
import librosa

# Set the path to your WAV file
wav_file = '/content/e.wav'

# Load the audio file using librosa.load()
y, sample_rate = librosa.load(wav_file, sr=None)

# Print the shape of the audio data array and the sample rate
print('Audio data shape:', y.shape)
print('Sample rate:', sample_rate)
```

```python
file_name='/content/e.wav'

y, sampling_rate = librosa.load(file_name)
librosa.display.waveshow(y,sr=sampling_rate)
ipd.Audio(file_name)
```

```python
pd.Series(y).plot(figsize=(10, 5),
                  lw=1,
                  title='Raw Audio Example',
                 color=color_pal[0])
plt.show()
```

```python
y_trimmed, _ = librosa.effects.trim(y, top_db=20)
pd.Series(y_trimmed).plot(figsize=(10, 5),
                  lw=1,
                  title='Raw Audio Trimmed Example',
                 color=color_pal[1])
plt.show()
```

```python
pd.Series(y[30000:30500]).plot(figsize=(10, 5),
                  lw=1,
                  title='Raw Audio Zoomed In Example',
                 color=color_pal[2])
plt.show()
```
